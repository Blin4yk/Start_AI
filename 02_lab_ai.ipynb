## 8.
1. Violinplot
Violinplot отображает распределение данных и одновременно позволяет увидеть медианы и квартили, как в boxplot. Однако дополнительно он показывает плотность значений, что может быть полезным.

2. Stripplot
Stripplot показывает каждую точку данных, распределенную по категориям целевой переменной. Это подходит для небольших или средних наборов данных.

3. Swarmplot
Swarmplot — улучшенная версия stripplot, где точки располагаются так, чтобы избежать перекрытия.

4. Pointplot
Pointplot отображает средние значения с доверительными интервалами, что может быть полезным для анализа трендов.

5. Barplot
Barplot также показывает средние значения, но вместо точек используются столбцы. Может быть полезен, если важен акцент на среднем значении.

6. Pairplot (для парных зависимостей)
Если вы хотите увидеть распределение признаков между собой или зависимости с целевой переменной, можно использовать pairplot

```python
import seaborn as sns

# Выбираем 5 признаков для построения графиков зависимости от целевой переменной
features = ['Age', 'Perseverance', 'Competitiveness', 'SelfReliance', 'SelfConfidence']

# Построение графиков зависимости признаков от целевой переменной
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(x='y', y=feature, data=df, ax=axes[i])
    axes[i].set_title(f'{feature} vs y')
    

plt.tight_layout()
plt.show()
```
